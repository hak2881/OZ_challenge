```python
# @title
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random

class HangmanGame:

    # í•„ìš”í•œ ê²ƒ
    # 1. í–‰ë§¨ì˜ í˜„ì¬ìƒíƒœ í‘œì‹œ
    # 2. ë‹¨ì–´ ì…ë ¥ë°›ê¸°
    # 3. ë‹¨ì–´ê°€ ë§ë‚˜ ì•ˆë§ë‚˜ í™•ì¸í•˜ê¸° ë§ìœ¼ë©´ í‘œê¸° í‹€ë¦¬ë©´ _
    ## í–‰ë§¨ìƒíƒœ ì¶œë ¥
    ## í˜„ì¬ ë‹¨ì–´ìƒíƒœ ì¶œë ¥
    ## ì¶”ì¸¡ë‹¨ì–´ ì²˜ë¦¬
    ##ê²Œì„ ì¢…ë£Œ ì—¬ë¶€

    def __init__(self):
        self.words = ["apple", "banana", "orange", "grape", "lemon"]
        self.turns = 7
        random.shuffle(self.words)
        self.word = self.words[0]
        self.fail =0
        self.guesses = ''
        self.hangman_pics = [
        """
        ------
        |    |
        |
        |
        |
        |
        ---
        """,
        """
        ------
        |    |
        |    O
        |
        |
        |
        ---
        """,
        """
        ------
        |    |
        |    O
        |    |
        |
        |
        ---""",
        """
        ------
        |    |
        |    O
        |   /|
        |
        |
        ---""",
        """
        ------
        |    |
        |    O
        |   /|\\
        |
        |
        ---
        """,
        """
        ------
        |    |
        |    O
        |   /|\\
        |   /
        |
        ---
        """,
        """
        ------
        |    |
        |    O
        |   /|\\
        |   / \\
        |
        ---
        """,
    ]

    ## í˜„ì¬ ë‹¨ì–´ìƒíƒœ ì¶œë ¥
    ## ì¶”ì¸¡ë‹¨ì–´ ì²˜ë¦¬

    # í–‰ë§¨ì˜ ìƒíƒœ ë³´ê¸°
    def display_hangman(self):
        print("í˜„ì¬ í–‰ë§¨ì˜ ìƒíƒœì—ìš”!")
        print(self.hangman_pics[7-self.turns])

    # ì–¼ë§ˆë‚˜ ë§ë‚˜ í™•ì¸í•˜ê¸°
    def check_letter(self):
        for chr in self.word :
                if chr in self.guesses :
                    print(chr,end=' ')
                else:
                    print('_',end=' ')
                    self.fail += 1

    # ì¶”ë¦¬í•œ ì•ŒíŒŒë²³ì´ ì•ˆì— ë“¤ì–´ê°€ë‚˜ í™•ì¸í•˜ê¸°
    def play(self):
        print('ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤!!.')
        print('ì•„ë˜ì˜ ê¸€ì ìˆ˜ì— ìœ ì˜í•˜ì—¬ ì¶”ì¸¡í•´ì„œ í–‰ë§¨ì„ ë ¤ì£¼ì„¸ìš”!')
        while self.turns > 0 :

            # ì¶”ì¸¡ ë‹¨ì–´ê°€ ë‹¨ì–´ì•ˆì— ìˆëŠ”ì§€ í™•ì¸
            self.check_letter()
            print()
            print()

            if self.fail == 0:
                break
            guess = input('ì¶”ì¸¡í•  ì•ŒíŒŒë²³ì„ ì…ë ¥í•˜ì„¸ìš”. ')
            # ë‹¤ìŒ ì¶”ë¦¬í•  ë¬¸ì ë§Œë“¤ê¸°
            self.guesses += guess

            #ëª‡ë²ˆ ê¸°íšŒê°€ ë‚¨ì•˜ëŠ”ì§€ ìˆ˜ë¦½
            print('ë‚¨ì€ ê¸°íšŒëŠ” {}ë²ˆ ì…ë‹ˆë‹¤'.format(self.turns-1))
            self.display_hangman()

            self.turns -= 1
            if self.turns ==0  :
                print ('You lose')




if __name__=='__main__':t

    game = HangmanGame()
    game.play()
```
